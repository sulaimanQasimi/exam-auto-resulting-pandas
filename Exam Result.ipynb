# Auto Exam Resulting

```python
import numpy as np
import pandas as pd
```

# Additional Settings

```python
# File Path
file="Exam.csv"
# Success Marks
success_mark=16
# read File
exam=pd.read_csv(file)
```

```python
# Get weither Student is PASS the exam or Not according to the success mark 
# And Create Result column and store the result of the opration 
exam["Result"]=np.where(exam["Score"]<16,"Success","Fail")
```

```python
# sort the result according to name and score 
final = exam.sort_values(by=['name','Score'],ascending=[True,False])
```

```python
# Export result to CSV
final.to_csv("Final_Exam_Result.csv")
# Export result to CSV
final.to_excel("Final_Exam_Result.xlsx")
```
